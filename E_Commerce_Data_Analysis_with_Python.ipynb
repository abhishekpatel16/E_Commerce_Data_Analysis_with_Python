There are 4 source systems in this project from where we need to load data in python data frames.

All these source systems are some third party databases, and have similar schema. For one of the source that is CSV_1, we also got source schema extract to understand the fields definition.

Note- Same schema is applicable for all 4 source systems datasets.

Each data source system contains 3 types of datasets named Customers, Products and Transaction.

Usecase-1 Load all three csv datasets in Pandas DataFrames and display first 5 records.


```python
import pandas as pd
```

```python
cust1 = pd.read_csv('customers.csv',names = ['custno','firstname','lastname','gender','age','profession','contactNo','emailId','city','state','isActive','createdDate','UpdateDate'])
prod1 = pd.read_csv('products.csv',names = ['productno','productName','Description','isActive','createdDate','UpdatedDate'])
tran1 = pd.read_csv('transactions.csv',names = ['txnno','txndate','custno','amount','productno','spendby'])
```

Usecase-2 Display only those customers from CSV_s1, who purchased more than 3 products.

```python
prodcount = tran1.groupby('custno')['productno'].count()
prodcount1 = prodcount.reset_index()
prodcount1[prodcount1['productno']>3]
```

```python
prodcount = tran1.groupby('custno')['productno'].count()
prodcount1 = prodcount.reset_index()
prodcount1.sort_values('productno', ascending = False)
#prodcount1[prodcount1['productno']>3]
```

```python
cust1.info()
```

```python
tran1.info()
```

```python
merged = cust1.merge(tran1, on = 'custno', how = 'right')
merged.info()
```

```python
merged = cust1.merge(tran1, on = 'custno', how = 'inner')
merged.info()
```

```python
prodcount = merged.groupby(['custno','firstname'])['productno'].count()
prodcount1 = prodcount.reset_index()
prodcount1.sort_values('productno', ascending = False)
#prodcount1[prodcount1['productno']>3]
```

Usecase-3 Display top 5 most demanded products from CSV_s1

```python
data = pd.merge(tran1,cust1[['custno','firstname','lastname']], on = 'custno')
data1 = pd.merge(data,prod1[['productno', 'productName']], on = 'productno')
product_demand = data1['productName'].value_counts().nlargest(5).reset_index()
product_demand.columns =['product_name','Demand']
product_demand
```
